```python
# Separate by EFI category
none_df = NP[NP["efi.strength"] == "None"].copy()
strong_weak_df = NP[NP["efi.strength"] != "None"].copy()
```

```python
# For EFI = None â†’ take the last entry per subject
none_last = (
    none_df.sort_values(by=["subject_id", "infant_age"])
    .groupby("subject_id", as_index=False)
    .last()
)
```

```python
#take the first entry where ct is numeric
strong_weak_df["ct_mean"] = pd.to_numeric(strong_weak_df["ct_mean"], errors="coerce")

sw_first = (
    strong_weak_df[strong_weak_df["ct_mean"].notna()]
    .sort_values(by=["subject_id", "infant_age"])
    .groupby("subject_id", as_index=False)
    .first()
)
sw_first.head(10)
```

```python
# Combine both
final = pd.concat([none_last, sw_first], ignore_index=True)
final = final.sort_values(by="subject_id").reset_index(drop=True)
```

```python
print(f"Subjects processed: {final.shape[0]}")
display(final.head())
```

```python
efi_strong = final[final["efi.strength"] == "Strong"]

counts = {
    "Unvaccinated": efi_strong["Unvaccinated"].sum(),
    "Recently Vaccinated": efi_strong["Recently Vaccinated"].sum(),
    "Vaccinated": efi_strong["Vaccinated"].sum()
}

print(counts)
```

```python
efi_weak_strong = final[final["efi.strength"].isin(["Weak", "Strong"])]

counts = {
    "Unvaccinated": efi_weak_strong["Unvaccinated"].sum(),
    "Recently Vaccinated": efi_weak_strong["Recently Vaccinated"].sum(),
    "Vaccinated": efi_weak_strong["Vaccinated"].sum()
}

print(counts)
```

```python
efi_strong_minimal = final[
    (final["efi.strength"] == "Strong") &
    (final["fsymptom"] == "Minimal")
]

counts = {
    "Unvaccinated": efi_strong_minimal["Unvaccinated"].sum(),
    "Recently Vaccinated": efi_strong_minimal["Recently Vaccinated"].sum(),
    "Vaccinated": efi_strong_minimal["Vaccinated"].sum()
}

print(counts)
```

```python
final['fsymptom'] = pd.Categorical(
    final['fsymptom'],
    categories=['None', 'Minimal', 'Moderate/Severe'],
    ordered=True
)
final['efi.strength'] = pd.Categorical(
    final['efi.strength'],
    categories=['None', 'Weak', 'Strong'],
    ordered=True
)
```

```python
# Extract only infants older than 2 months
pmd3_2plus = pmd3[pmd3["age_group"] == "2+ months"].copy()

# Create a single vaccination status column
def get_vaccination_status(row):
    if row["Vaccinated"]:
        return "Vaccinated"
    elif row["Recently Vaccinated"]:
        return "Recently Vaccinated"
    elif row["Unvaccinated"]:
        return "Unvaccinated"
    else:
        return "Unknown"

pmd3_2plus["vaccination_status"] = pmd3_2plus.apply(get_vaccination_status, axis=1)

# Drop old vaccination columns
pmd3_2plus = pmd3_2plus.drop(columns=["Unvaccinated", "Recently Vaccinated", "Vaccinated"])

# Preview
pmd3_2plus.head()
```

```python
# Create a clean index for the filtered dataset
pmd3_2plus.reset_index(drop=True, inplace=True)
pmd3_2plus.index = pmd3_2plus.index + 1
# Save to Excel with the new index
pmd3_2plus.to_excel("pmd3_2plus.xlsx", index=True)
```

```python
age_table = pd.DataFrame({
    "Unvaccinated": [
        neonates_pmd3["Unvaccinated"].sum(),
        onetwo_pmd3["Unvaccinated"].sum(),
        twoplus_pmd3["Unvaccinated"].sum()
    ],
    "Recently Vaccinated": [
        neonates_pmd3["Recently Vaccinated"].sum(),
        onetwo_pmd3["Recently Vaccinated"].sum(),
        twoplus_pmd3["Recently Vaccinated"].sum()
    ],
    "Vaccinated": [
        neonates_pmd3["Vaccinated"].sum(),
        onetwo_pmd3["Vaccinated"].sum(),
        twoplus_pmd3["Vaccinated"].sum()
    ]
}, index=["Neonates", "1â€“2 months", "2+ months"])

# Add sum column
age_table["Total"] = age_table.sum(axis=1)

print("age of first infection by vaccine status:")
display(age_table)
```

```python
# Total number of first infections
n_total = 177

# Calculate proportions
age_prop = (age_table / n_total * 100).round(1)

print("Proportion (%) of first infections by age group and vaccine status:")
display(age_prop)
```

```python
age_prop_combined = age_table.copy()
for col in age_table.columns:
    age_prop_combined[col] = (
        age_table[col].astype(str) + " (" + (age_table[col] / n_total * 100).round(1).astype(str) + "%)"
    )

print("First infection by age group and vaccine status (count and %):")
display(age_prop_combined)
```
